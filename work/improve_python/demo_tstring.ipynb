```python
# Templateオブジェクトのインポート
from string.templatelib import Template
```

```python
name = "Python"
# f-string: すぐに "Hello Python" という文字列になる
f_res = f"Hello {name}" 

# t-string: 構造を保持したオブジェクトになる
t_res = t"Hello {name}"

print(type(t_res)) # <class 'string.templatelib.Template'>
print(t_res.strings) # ('Hello ', '')
print(t_res.values)  # ('Python',)
```
